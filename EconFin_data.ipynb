```python
import time
from pytrends.request import TrendReq
import pandas as pd

# Crear una instancia de TrendReq
pytrends = TrendReq(hl='es-ES', tz=360)
time.sleep(5)
# Definir las palabras clave y el período de tiempo
keywords = ['Toyota', 'Dolares', 'Crisis', 'Inflación']
timeframe = '2000-01-01 2024-01-01'  # Formato: 'YYYY-MM-DD YYYY-MM-DD'

# Construir el payload de la consulta
pytrends.build_payload(keywords, cat=0, timeframe='all', geo='BO', gprop='')
#pytrends.build_payload(keywords, cat=0, timeframe='today 5-y', geo='', gprop='')
# Obtener los datos de interés a lo largo del tiempo
interest_over_time_df = pytrends.interest_over_time()
time.sleep(5)
df = pd.DataFrame(interest_over_time_df)

interest_over_time_df_ = pytrends.interest_over_time()
monthly_data = interest_over_time_df_.resample('M').sum()
df1_montly = pd.DataFrame(monthly_data)
###
time.sleep(10)
keywordsT = ['Dolares']
pytrends.build_payload(keywordsT, cat=0, timeframe='all', geo='BO', gprop='')
interest_over_time_dfT = pytrends.interest_over_time()
dfT = pd.DataFrame(interest_over_time_dfT)
###

time.sleep(15)
keywords1 = ['Computadora', 'Viajes', 'Tipo de cambio']
pytrends.build_payload(keywords1, cat=0, timeframe='all', geo='', gprop='')
interest_over_time_df1 = pytrends.interest_over_time()
df1 = pd.DataFrame(interest_over_time_df1)
```

```python
# Guardar DataFrame en un archivo 'name.xlsx' en tu carpeta de Google Drive.
ruta_archivo = '/content/drive/MyDrive/Google_trends/resultados.xlsx'
#df.to_excel(ruta_archivo, sheet_name="example", index=False)  # El argumento index=False evita que se guarde el índice en el archivo Excel.
df.to_excel(ruta_archivo, sheet_name="example")
# Carga el archivo Excel existente.
with pd.ExcelWriter(ruta_archivo, mode='a', engine='openpyxl') as writer:
    df1.to_excel(writer, sheet_name="example1")
    df1_montly.to_excel(writer, sheet_name="example1_mon")
    dfT.to_excel(writer, sheet_name="EXAMPLE")
    #df1.to_excel(writer, sheet_name="inflación divisiones", index=False)
```

```python
import plotly.graph_objs as go
# Crear un gráfico de líneas para cada variable
fig = go.Figure()

# Agregar cada variable al gráfico
for keyword in keywords:
    fig.add_trace(go.Scatter(x=interest_over_time_df.index, y=interest_over_time_df[keyword], mode='lines', name=keyword))

# Establecer el diseño del gráfico
fig.update_layout(
    title='Tendencias de Búsqueda en Google',
    xaxis_title='Fecha',
    yaxis_title='Interés',
    template='plotly_dark'
)

# Mostrar el gráfico
fig.show()
```
